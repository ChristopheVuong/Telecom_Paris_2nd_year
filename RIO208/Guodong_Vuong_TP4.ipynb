# Generation of a set of vertices

We generate vextices based on poisson process. Then find the connected edge and triangle. 

```python
def proc_poisson(lbda, a=1):
    # The points set is the vertexes
    N = np.random.poisson(lbda*a**2)
    L_0 = a * np.random.rand(N, 2)
    return L_0

def simplex_1(simplex0, r):
    # The L_1 set are edges which 
    L_1 = []
    n = len(simplex0)
    for i in range(n):
        for j in range(i + 1, n):
            if np.linalg.norm(simplex0[j, :] - simplex0[i, :]) <= r:
                L_1.append([i, j]) # the list is an index list, here we are.
    return L_1

def simplex_2(simplex0 ,simplex1):
    L_2 = []
    N = len(simplex0)
    for u in simplex1:
        for i in range(u[1], N):
            if [u[0], i] in simplex1 and [u[1], i] in simplex1:
                L_2.append([u[0], u[1], i])
    return L_2
```

```python
# simplex_0 is vertex list, simplex_1 is edge list, simplex_2 is triangle list. 
list_dim0 = proc_poisson(100)
plt.scatter(list_dim0[:,0], list_dim0[:, 1])
```

# Calcul de la topologie

```python
# r the radius of the circles
def delta1(list_0, list_1):
    cols = len(list_1)
    rows = len(list_0)
    matrix = np.zeros((rows, cols))
    for i in range(cols):
        edge = list_1[i]
        matrix[edge[0],i] = -1
        matrix[edge[1],i] = 1
    return matrix

def delta2(list_1, list_2):
    cols = len(list_2)
    rows = len(list_1)
    matrix = np.zeros((rows, cols))
    for i in range(cols):
        triangle = list_2[i]
        matrix[list_1.index([triangle[0], triangle[1]]),i] = 1
        matrix[list_1.index([triangle[0], triangle[2]]),i] = -1
        matrix[list_1.index([triangle[1], triangle[2]]),i] = 1
    return matrix
```

```python
# The delta1 indicate the transition from vertex (0-simplex) to 1-simplex. 
transition_1 = delta1(list_dim0, list_dim1)
transition_1
```

```python
# The delta1 indicate the transition from 1-simplex to 2-simplex. 
transition_2 = delta2(list_dim1, list_dim2)
transition_2
```

```python
def betti0(list_0, transition_1):
    betti_0 = len(list_0) - np.linalg.matrix_rank(transition_1)
    return betti_0
def betti1(transition_1, transition_2):
    betti_1 = (transition_1.shape[1]- np.linalg.matrix_rank(transition_1)) - np.linalg.matrix_rank(transition_2)
    return betti_1
```

```python
betti_0 = betti0(list_dim0, transition_1)
betti_1 = betti1(transition_1, transition_2)
print('The number of connected vextexes 0 is ', betti_0, ' the number of coverage hole', betti_1)
```

```python
# The Euler characteristics:
def euler_charater_betti(betti_0, betti_1):
    eu = betti_0-betti_1
    return eu
```

```python
a = euler_charater_betti(betti_0, betti_1)
print('The euler charateristic number is',a)
```

# Meanfield Results

```python
def session(lbda):
    # for each lambda, we run 200 session:
    session_data = np.zeros((100,3))
    for j in range(100):
        vertex = proc_poisson(lbda, a=1)
        edge = simplex_1(vertex, 0.1)
        triangle = simplex_2(vertex ,edge)
        if len(vertex) == 0 or len(edge) == 0 or len(triangle) == 0:
            continue
        transition_1 = delta1(vertex, edge)
        transition_2 = delta2(edge, triangle)
        
        session_data[j,0] = betti0(vertex, transition_1)
        session_data[j,1] = betti1(transition_1, transition_2)
        session_data[j,2] = session_data[j,0] - session_data[j,1]
    session_data_mean = np.mean(session_data, axis = 0)
    return session_data_mean
```

```python
data_mean = np.zeros((200, 3))
# the first dimension represent each 
for i in range(200):
    lbda = i+1
    if i%5==0:
        print(i)
    data_mean[i,:] = session(lbda)
```

```python
# plot the betti1 in terms of lambda
plt.figure
plt.plot(range(1,201), data_mean[:,0])
plt.xlabel('lambda')
plt.ylabel('la nombre de betti 1')
plt.show()
```

```python
# plot the betti2 in terms of lambda
plt.figure
plt.plot(range(1,201), data_mean[:,1])
plt.xlabel('lambda')
plt.ylabel('la nombre de betti 2')
plt.show()
```

```python
# plot the euler number in terms of lambda
plt.figure
plt.plot(range(1,201), data_mean[:,2])
plt.xlabel('lambda')
plt.ylabel('la caractéristique d’Euler')
plt.show()
```
